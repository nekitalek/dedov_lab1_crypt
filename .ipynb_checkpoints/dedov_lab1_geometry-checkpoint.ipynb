```python
import random
from sympy import *

def check_D(D):
    if D % 8 == 1 and D % 3 != 0:
        return 1
    else:
        return 0
    
def choose_D():
    for i in range(-11000000, -10000000):
        if check_D(i) == 1:
            return i
        
def is_prime(p):
    for i in range(2, int(p/2)):
        if(p % i) == 0:
            return False
    return True

def find_a_b_method1(D):
    D *= -1
    while(1):
        a = random.randint(2**123, 2**124)
        b = random.randint(2**123, 2**124)
        p = ((D + 1) // 4)*(a * a) + D * a * b + D * b * b
        print("a, b, p = ", a, b, p)
        if(isprime(p)):
            print("p is prime")
            c = 3
            N1 = p + 1 + a
            N2 = p + 1 - a
            N1 /= c
            N2 /= c
            if (N1 >= 2**252 or N2 >= 2**252):
                print("find is end: a, b, p = ", a, b, p)
                return a, b
            
def find_a_b(D):
    a, b = 0, 0
    if D % 4 == 1:
        a, b = find_a_b_method1(D)
        
#D = -10999999
D = choose_D()
print("Discriminate =", D)
K = QuadraticField(D)
H_d = K.class_number()
print("H_d =", H_d)
find_a_b(D)
```
